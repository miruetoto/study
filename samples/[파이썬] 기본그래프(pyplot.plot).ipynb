# 7. 마커의 종류를 바꾸고 싶다. R에서 `pch`, `lty` 등을 통합

점선을 그리고 싶다면 하고싶다면 `:`을 사용한다. 즉 
```
plot(X1,':')
```
이라고 하면 된다. 점선과 함께 스캐터를 그리고 싶다면 

````python
plot(X1,':o',color='blue')
plot(X2,':o',color='red')
````
